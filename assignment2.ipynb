```
install.packages("SportsAnalytics")
library(SportsAnalytics)
NBA1516<-fetch_NBAPlayerStatistics("15-16")
```

```
install.packages("ggplot2")
```

```
library(SportsAnalytics)
library(ggplot2)

qplot(ThreesAttempted,ThreesMade,data=NBA1516)

qplot(ThreesAttempted,ThreesMade,data=NBA1516,geom=c("point", "smooth"))

qplot(ThreesAttempted,data=NBA1516,fill=ThreesMade)

qplot(ThreesAttempted,ThreesMade,data=NBA1516,facets = Position~.)
```

```
library(SportsAnalytics)
library(ggplot2)

qplot(GamesPlayed,FieldGoalsMade,data=NBA1516)

qplot(GamesPlayed,FieldGoalsMade,data=NBA1516,geom=c("point", "smooth"))

qplot(GamesPlayed,data=NBA1516,fill=FieldGoalsMade)

qplot(GamesPlayed,FieldGoalsMade,data=NBA1516,facets = Position~.)
```
