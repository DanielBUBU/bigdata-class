```
NBA1516 %>%
  count(Position) %>%
  ggplot(aes(x=fct_reorder(Position,n),y=n,fill=Position))+
  geom_col()+
  geom_text(aes(label=n),vjust=-0.25)+
  labs(title="站位數量",x="站位",y="數量")+
  theme(legend.position = "none")
```
