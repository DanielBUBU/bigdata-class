```
library(readr)
library(ggplot2)
dataset <- read.csv("PM2.5.csv")
```

```
install.packages("showtext")
library(showtext)
showtext.auto(enable = TRUE)
```

```
dataset
```

```
ggplot(data = dataset,aes(PM25,	county))+geom_point()
ggplot(data = dataset,aes(PM25,	county))+geom_boxplot()
```

```
qplot(PM25,data=dataset,facets = county~.,binwidth=0.5)
qplot(PM25,data=dataset,facets = county~.,binwidth=10)
qplot(PM25,data=dataset,facets = county~.,binwidth=100)
```

```
qplot(PM25,data=dataset,fill=county)
```

```
ggplot()+geom_bar(data = dataset,aes(x=PM25,y=county),stat = "identity")
```
