```
eletricty_data <- read.csv("台灣電力公司_太陽光電發電量及平均單位裝置容量每日發電量統計表.csv")
eletricty_data
```

月分與平均單位裝置容量每日發電量的關係???

```
ggplot(eletricty_data,aes(x=month	,y=Average.unit.capacity.per.day))+geom_point()+geom_smooth()
```

Average.unit.capacity.per.day與month的關係且用year來區隔顏色???

```
ggplot(eletricty_data,aes(x=month	,y=Average.unit.capacity.per.day,color=year))+geom_line()+facet_grid(year~.)
```

month和Average.unit.capacity.per.day用boxplot繪出??

```

ggplot(eletricty_data,aes(x=month	,y=Average.unit.capacity.per.day,fill=year))+geom_boxplot()
```

每日發電量分布

```
qplot(Average.unit.capacity.per.day,data=eletricty_data)
```

Average.unit.capacity.per.day和month用geom_point()和geom_smooth()顯示關係??

```
ggplot(eletricty_data,aes(x=month	,y=Average.unit.capacity.per.day))+geom_point()+facet_grid(year~.)+geom_smooth()
```
